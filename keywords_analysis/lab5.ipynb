# Лабораторная работа 5 #

## Выполнила: Иванова Дарья Сергеевна БПМ-23-ПО-2 

Задача: Необходимо определить ключевые слова в текстах политических статей с помощью TF-IDF и оценить, насколько часто эта лексика используется в комментариях. Для каждой статьи выделяются наиболее значимые слова и словосочетания, после чего считается их частота в комментариях к соответствующей статье. Дополнительно рассчитывается процент совпадения каждого комментария с ключевым словарём статьи. На основе этих данных строится статистика по комментаторам: количество комментариев, число упоминаний ключевых слов и средний процент тематического совпадения. Это позволяет определить, какие пользователи обсуждают содержание статьи, а какие в основном оставляют оценочные или отвлечённые реплики
